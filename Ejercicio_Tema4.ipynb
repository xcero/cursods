# Resolviendo los problemas con los datos faltantes:

Reemplazar con la media o la mediana

La variable normalized-losses tiene 41 datos faltantes. Revisar la simetría del atributo y luego decidir si se utilza media o mediana:

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.distplot(df["normalized-losses"])
```

```python
mean = df['normalized-losses'].mean()
median = df['normalized-losses'].median()
mode = df['normalized-losses'].mode()
skew = df['normalized-losses'].skew()
kurt = df['normalized-losses'].kurt()
print("La media es:", mean)
print("La mediana es:", median)
print("La moda es:", mode)
print("El sesgo es:", skew)
print("La kurtosis es:", kurt)
```

```python
#Calcular el mediana de la columna
m_norm_loss = df["normalized-losses"].astype("float").median(axis=0)
#Remplazar "NaN por el valor de la mediana en la columna "normalized-losses"
df["normalized-losses"] = df["normalized-losses"].replace(np.nan, m_norm_loss)
```

```python
df.head(20)
```

Contar Valores

Una forma de resumir los datos categóricos es usando la función value_counts. Por ejemplo, en el conjunto de datos, se tiene el lugar del motor (engine-location) como una variable categórica de frontal y trasero.

```python
location_counts = df['engine-location'].value_counts().to_frame()
location_counts
```

Remplazar con la Frecuencia:

La variable num-of-doors tiene 2 datos faltantes, remplazar con "four". Razón: El 84% de los sedanes es de cuatro puertas. Debido que tener cuatro puertas es mas probable que ocurra

```python
door_count=df['num-of-doors'].value_counts().to_frame()
door_count
```

```python
#Reemplazamos los valores faltantes en 'num-of-doors' con el valor más frecuente o la moda
df["num-of-doors"]=df["num-of-doors"].replace(np.nan, "four")
```
